```python
import pandas as pd
import numpy as np
from sklearn.impute import KNNImputer

# Create a dummy dataset with missing values
data = {
    'Age': [25, np.nan, 30, 22, 40, np.nan, 28],
    'Salary': [50000, 60000, np.nan, 52000, 58000, 62000, np.nan],
    'City': ['New York', 'Los Angeles', 'New York', np.nan, 'Chicago', 'Chicago', 'Los Angeles'],
    'Purchased': ['Yes', 'No', np.nan, 'No', 'Yes', 'Yes', 'No']
}

df_dummy = pd.DataFrame(data)
print("Original Dummy Dataset with Missing Values:")
print(df_dummy)
```

1. Drop columns or rows with excessive missing values.

```python
threshold = 0.5
df_cleaned = df_dummy.loc[:, df_dummy.isnull().mean() < threshold]
df_cleaned = df_cleaned.dropna(thresh=int((1-threshold)*len(df_cleaned)), axis=0)

print("\nAfter Dropping Excessive Missing Values:\n")
print(df_cleaned)
```

2. Impute missing numerical columns (Age, Salary) with mean or median.

```python
df_num = df_cleaned[['Age','Salary']].copy()
df_num['Age'] = df_num['Age'].fillna(df_num['Age'].median())
df_num['Salary'] = df_num['Salary'].fillna(df_num['Salary'].mean())

print("\nAfter Numerical Imputation (Median for Age, Mean for Salary):\n")
print(df_num)
```

3. Impute missing categorical columns (City, Purchased) with mode.

```python
df_cat = df_cleaned[['City','Purchased']].copy()
df_cat['City'] = df_cat['City'].fillna(df_cat['City'].mode()[0])
df_cat['Purchased'] = df_cat['Purchased'].fillna(df_cat['Purchased'].mode()[0])

print("\nAfter Categorical Imputation (Mode):\n")
print(df_cat)

df_final = pd.concat([df_num, df_cat], axis=1)

print("\nFinal Dataset after Simple Imputation:\n")
print(df_final)
```

4. Optionally, apply KNN imputation for numerical columns.

```python
imputer = KNNImputer(n_neighbors=2)
df_knn_num = pd.DataFrame(imputer.fit_transform(df_cleaned[['Age','Salary']]),
                          columns=['Age','Salary'])

print("\nAfter KNN Imputation for Numerical Columns:\n")
print(df_knn_num)
```
